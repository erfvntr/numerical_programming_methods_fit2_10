```python
# Метод Ейлера
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x, y):
    return x + np.sin(y / math.sqrt(7))
x0 = 0.5
x_end = 1.5
y0 = 0.6
h = 0.2

def euler_method(f, x0, y0, h, x_end):
    x_values = [x0]
    y_values = [y0]
    while x0 < x_end:
        y0 = y0 + h * f(x0, y0)
        x0 += h
        x_values.append(x0)
        y_values.append(y0)
    return x_values[:-1], y_values[:-1]  

x_values, y_values = euler_method(f, x0, y0, h, x_end)
for i in range(len(x_values)):
    x_rounded = round(x_values[i], 4)
    y_rounded = round(y_values[i], 4)
    plt.text(x_values[i], y_values[i] + 0.03, f"({x_rounded}, {y_rounded})", fontsize=8, ha='right')
    print(f"x = {x_rounded}, y = {y_rounded}")

plt.plot(x_values, y_values, "o-", color="red", label=" y' = x + sin(y/√7)") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Метод Ейлера") 
plt.grid()
plt.legend()
plt.show()

```

```python
#Розв'язок диф.рівняння через odeint
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def f(y, x):
    return x + np.sin(y / np.sqrt(7))
x0 = 0.5
x_end = 1.5
y0 = 0.6
h = 0.2

def model(y, x):
    dydx = f(y, x)
    return dydx

x_points = np.arange(x0, x_end + h, h)
y_solution = odeint(model, y0, x_points)

for i in range(len(x_points)):
    x_rounded = round(x_points[i], 4)
    y_rounded = round(y_solution[i][0], 4)
    plt.text(x_points[i], y_solution[i][0] + 0.03, f"({x_rounded}, {y_rounded})", fontsize=8, ha='right')
    print(f"x = {x_rounded}, y = {y_rounded}")

plt.plot(x_points, y_solution[:, 0], "o-", color="blue", label=" y' = x + sin(y/√7) ") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Розв'язок за допомогою scipy.odeint") 
plt.grid()
plt.legend()
plt.show()
```

```python
# Метод Ейлера-Коші
import numpy as np
import matplotlib.pyplot as plt

def euler_cosh(f, x0, y0, h, x_end):
    x_values = [x0]
    y_values = [y0]
    while x0 < x_end:
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + h, y0 + k1)
        y0 = y0 + 0.5 * (k1 + k2)
        x0 += h
        x_values.append(x0)
        y_values.append(y0)
    return x_values, y_values

def f(x, y):
    return x + np.cos(y/np.exp(x))  

x0 = 1.4 
y0 = 2.5
h = 0.2 
x_end = 2.4 
x_values, y_values = euler_cosh(f, x0, y0, h, x_end)

for i in range(len(x_values)):
    x_rounded = round(x_values[i], 4)
    y_rounded = round(y_values[i], 4)
    plt.text(x_values[i], y_values[i]+0.03, f"({x_rounded}, {y_rounded})", fontsize=8, ha='right')
    print(f"x = {x_rounded}, y = {y_rounded}")

plt.plot(x_values, y_values, "o-", color="blue", label="y' = x + cos(y/e)") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Метод Ейлера-Коші") 
plt.grid()
plt.legend()
plt.show()
```
